Criando faixa-etárias para classificar a sociedade de acordo com a idade:

```python
def faixasidade(df):
    x=df['V8005']
    if x<10: return 1
    elif x<=30: return 2
    elif x<60: return 3
    elif x>=60: return 4
    return 5
faixas=dados.apply(faixasidade,axis=1)
df['faixas']=faixas
df.faixas=df.faixas.astype('category')
df.faixas.cat.categories=(['Crianças','Jovens','Adultos','Idosos'])
```

*As faixas etárias foram criadas pela minha percecpção do conceito de criança e jovens

```python
df.head()
```

Criando as regiões do Brasil:

```python
def regiao(df):
    es=df['UF']
    if es <= 17: return 1
    elif es <= 29: return 2
    elif es <= 35: return 3
    elif es <= 43: return 4
    elif es >= 50: return 5
    return 6
reg=dados.apply(regiao,axis=1)
df['Regiaobr'] = reg
df.Regiaobr=df.Regiaobr.astype('category')
df.Regiaobr.cat.categories=(['Norte','Nordeste','Sudeste','Sul','Centro-Oeste'])
```

Faixa etária existente em cada região do Brasil:

```python
ct2 = pd.crosstab(df['Regiaobr'],df['faixas'], normalize=True)

plot = ct2.plot(kind='bar', stacked = True)
plt.ylabel('População')
plt.xlabel('Região')
plt.show()
```

    Aqui pode-se observar que em todas as regiões encontra-se uma maior quantidade de adultos.

```python
#crosstab entre se tem telefone pessoal e regiao  do brasil
ct3 = pd.crosstab(df['Tem telefone móvel celular para uso pessoal'], df['Regiaobr'], normalize= 'columns')
ct3
```

```python
plot = ct3.T.plot(kind='bar',figsize=(6,6), stacked = True)
plt.ylabel('População')
plt.xlabel('Região brasileira')
plt.title('População por região brasileira que possui ou não telefone par uso pessoal(1)')
plt.show()
```

    Nas regiões Sudeste, Sul e Centro-Oeste mais de 80% de seus habitantes possuem um telefone móvel para uso pessoal. Nas demais regiões mais de 70% possuem o telefone celular. Nos gráficos abaixo é possível visualizar cada região, com maoir precisão.

Cada unidade de federação de maneira individual para melhor percepção:

```python
re =  pd.crosstab(df['Tem telefone móvel celular para uso pessoal'], df.Regiaobr[df.Regiaobr == 'Norte'],normalize=True)
rep = re.plot(kind='bar', stacked = True, color = 'pink')
plt.ylabel('Porcentagem populacional')
plt.title('População no norte que possui ou não telefone par uso pessoal(2)')
plt.show()
 
re2 =  pd.crosstab(df['Tem telefone móvel celular para uso pessoal'], df.Regiaobr[df.Regiaobr == 'Sul'],normalize=True)
rep2 = re2.plot(kind='bar', stacked = True, color = 'purple')
plt.ylabel('Porcentagem populacional')
plt.title('População no sul que possui ou não telefone par uso pessoal(3)')
plt.show()

re3 =  pd.crosstab(df['Tem telefone móvel celular para uso pessoal'], df.Regiaobr[df.Regiaobr == 'Centro-Oeste'],normalize=True)
rep3 = re3.plot(kind='bar', stacked = True, color = 'green')
plt.ylabel('Porcentagem populacional')
plt.title('População no Centro=-Oeste que possui ou não telefone par uso pessoal(4)')
plt.show()

re4 =  pd.crosstab(df['Tem telefone móvel celular para uso pessoal'], df.Regiaobr[df.Regiaobr == 'Nordeste'],normalize=True)
rep4 = re4.plot(kind='bar', stacked = True, color = 'red')
plt.ylabel('Porcentagem populacional')
plt.title('População no nordeste que possui ou não telefone par uso pessoal(5)')
plt.show()

re5 =  pd.crosstab(df['Tem telefone móvel celular para uso pessoal'], df.Regiaobr[df.Regiaobr == 'Sudeste'],normalize=True)
rep5 = re5.plot(kind='bar', stacked = True, color = 'orange')
plt.ylabel('Porcentagem populacional')
plt.title('População pno sudeste que possui ou não telefone par uso pessoal(6)')
plt.show()
```
