Criando meu data frame:

Criando faixa-etárias para classificar a sociedade de acordo com a idade:

```python
def faixasidade(df):
    x=df['V8005']
    if x<10: return 1
    elif x<=30: return 2
    elif x<60: return 3
    elif x>=60: return 4
    return 5
faixas=dados.apply(faixasidade,axis=1)
df['faixas']=faixas
df.faixas=df.faixas.astype('category')
df.faixas.cat.categories=(['Crianças','Jovens','Adultos','Idosos'])
```

*As faixas etárias foram criadas pela minha percecpção do conceito de criança e jovens

```python
df.head()
```

Criando as regiões do Brasil:

```python
def regiao(df):
    es=df['UF']
    if es <= 17: return 1
    elif es <= 29: return 2
    elif es <= 35: return 3
    elif es <= 43: return 4
    elif es >= 50: return 5
    return 6
reg=dados.apply(regiao,axis=1)
df['Regiaobr'] = reg
df.Regiaobr=df.Regiaobr.astype('category')
df.Regiaobr.cat.categories=(['Norte','Nordeste','Sudeste','Sul','Centro-Oeste'])
```

Faixa etária existente em cada região do Brasil:

```python
ct2 = pd.crosstab(df['Regiaobr'],df['faixas'], normalize=True)

plot = ct2.plot(kind='bar', stacked = True)
plt.ylabel('População')
plt.xlabel('Região')
plt.show()
```

    Aqui pode-se observar que em todas as regiões encontra-se uma maior quantidade de adultos.

## Conclusão 

    Pelo que a PNAD de 2015 apontou, não houve nenhuma criança (de 0 a 10 anos), que tivesse seu próprio telefone móvel para uso pessoal, por mais que se veja várias crianças, de 8 a 10 anos já possuem um celular. Os adultos são os que mais possuem um celular para uso pessoal, em seguda os jovens e por último os idosos. No entanto, dado a quantidade de idosos que respoderam à PNAD, a quantidade de anciãos que estão se mordenizando é muito significativa. Assim, a maioria acessa a internet via telefone celular, a poucos por meio de tablet. O mesmo ocorre para os jovens e adultos. Isso se dá pela praticidade do smartphone, em que é possível realizar tudo, o que antes era feito em computadores.
    A quantidade de idosos no Sudeste é maior que nas demais unidades de federações, esse pode ser um dos motivos de haver outliers na região, já que os idosos são os que possuem maior renda mensal, e esse foi o Estado com maior número de entrevistados.
    Mais mulheres foram entrevistadas do que homems, o que pode ser um dos motivos de a renda masculina ter dado maior que a feminina. No entanto, é de conhecimento comum que há uma diferença entre o salário de um homem e de uma mulher, em que o homem ganha mais, independente de seu cargo. Com isso, essa informaçao da PNAD, pode ser a ilustração veradeira da desigualdade salarial no Brasil. Contudo, ambos sexos possuem uma maior porcentagem que utiliza o telefone celular para acessar a internet, e que possuem um celular para uso pessoal.
    Nas regiões onde a renda é maior, há maior freuqência de celular para uso pessoal, apesar da desigualdade social existente. Como no Cento-Oeste, em que uma quantidade de pessoas foram entrevistadas, mas que mostrou possuir a maior renda concentrada no DF, local que possui grande desigualdade social. O mesmo ocorre com o Sudeste, porém mais pessoas responderam ao questionário.
    Assim, todas as gerações acessam à internet, majoritariamente, por meio do telefone celular, pouco por tablets e quase nada por outro tipo de equipamento eltrônico. Isso já é uma tendência que se vêm acompanhando ao longo dos tempos. Dessa forma, a geração idosa vem sim se mordenizando, de tal forma que possuem celulares próprios e acessam a internet através deles. Enquanto os jovens são os maiores usuários de celulares para se conectarem à internet. O que também não é de se estranhar, visto que foi a geração que cresceu ao lado das invações tecnológicas que foram introduzidas à sociedade no século XXI. Por mais que exista uma diferença de renda entre Estados, Regiões e sexos, a grande maioria os  brasileiros, está englobada no mundo tecnológico e de alguma maneira, via telefone móvel e tablets, acessam a internet. Além de possuirem um telefone móvel para uso pessoal. Isso se tornará um número crescente dentro do Brasil.
